```python
import os
import cv2
```

```python
dirp = '/mnt/e/TUNEX/Augmentation/'
```

```python
import albumentations
```

### Using the [albumentations](https://albumentations-demo.herokuapp.com/) python library, we are going to perform the following transforms on the dataset :

### Rotation - 30 degrees (left and right)

```python
R30L = albumentations.Rotate(limit=(30, 30), interpolation=1, border_mode=4,  p=1)
R30R = albumentations.Rotate(limit=(-30, -30), interpolation=1, border_mode=4,  p=1)
```

### Rotation - 15 degrees (left and right)

```python
R15L = albumentations.Rotate(limit=(15, 15), interpolation=1, border_mode=4,  p=1)
R15R = albumentations.Rotate(limit=(-15, -15), interpolation=1, border_mode=4,  p=1)
```

### Horizontal Flip

```python
HFlip = albumentations.HorizontalFlip(always_apply=False, p=1.0)
```

### Random Brightness and Contrast

```python
Bright = albumentations.RandomBrightnessContrast(always_apply=False, p=1.0, brightness_limit=(-0.25, 0.23), contrast_limit=(-0.35, 0.18), brightness_by_max=True)
```

> Current Data
#### The [KDEF dataset](https://kdef.se/) have already been extracted and arranged according to the emotions, in the "data_in" directory.
#### Each emotion contains `420 sets of images`, by using only the straight, half-left and half-right faced images from the [KDEF dataset](https://kdef.se/).
#### Therefore, at this moment there are a total of `2,940 images`.
#### 
### We will now perform 6 different types of augmentation techniques as discussed, on this data. 
### And separate them in different directories, according to their transform type.
#### 

```python
count = 1
for n in range(0,7):
    dirt = dirp+'data_in/'+str(n)+'/'
    print(dirt)
    for x,y,z in os.walk(dirt):
        for i in z:
            img = cv2.imread(x+'/'+i)
            print('read!..  '+i)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            print('converted!')

            new = R30L(image=img)['image']
            new = cv2.cvtColor(new, cv2.COLOR_RGB2BGR)
            if not os.path.exists(dirp+'R30L/'+str(n)+'/'):
                        os.mkdir(dirp+'R30L/'+str(n)+'/')
            cv2.imwrite(dirp+'R30L/'+str(n)+'/R30L'+i, new)

            new = R30R(image=img)['image']
            new = cv2.cvtColor(new, cv2.COLOR_RGB2BGR)
            if not os.path.exists(dirp+'R30R/'+str(n)+'/'):
                        os.mkdir(dirp+'R30R/'+str(n)+'/')
            cv2.imwrite(dirp+'R30R/'+str(n)+'/R30R'+i, new)

            new = R15L(image=img)['image']
            new = cv2.cvtColor(new, cv2.COLOR_RGB2BGR)
            if not os.path.exists(dirp+'R15L/'+str(n)+'/'):
                        os.mkdir(dirp+'R15L/'+str(n)+'/')
            cv2.imwrite(dirp+'R15L/'+str(n)+'/R15L'+i, new)

            new = R15R(image=img)['image']
            new = cv2.cvtColor(new, cv2.COLOR_RGB2BGR)
            if not os.path.exists(dirp+'R15R/'+str(n)+'/'):
                        os.mkdir(dirp+'R15R/'+str(n)+'/')
            cv2.imwrite(dirp+'R15R/'+str(n)+'/R15R'+i, new)

            new = HFlip(image=img)['image']
            new = cv2.cvtColor(new, cv2.COLOR_RGB2BGR)
            if not os.path.exists(dirp+'HFlip/'+str(n)+'/'):
                        os.mkdir(dirp+'HFlip/'+str(n)+'/')
            cv2.imwrite(dirp+'HFlip/'+str(n)+'/HFlip'+i, new)

            new = Bright(image=img)['image']
            new = cv2.cvtColor(new, cv2.COLOR_RGB2BGR)
            if not os.path.exists(dirp+'Bright/'+str(n)+'/'):
                        os.mkdir(dirp+'Bright/'+str(n)+'/')
            cv2.imwrite(dirp+'Bright/'+str(n)+'/Bright_'+i, new)
            
            print('done...'+str(count))
            count += 1
```
